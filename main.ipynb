```python
## Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
```

DATA CLEANING

```python
## Missing values
data.isnull().sum()
```

```python
data[data.isnull().any(axis=1)]
```

The DataSet is converted into two sets based on Region from 122nd index, so we can make a new column based on the region
0: "Bejaia Region Dataset"
1: "Sidi-Bel Abbes Region Dataset" 

```python
data.loc[:122,"Region"] =  0
data.loc[122:,"Region"] =  1
```

```python
data["Region"] = data["Region"].astype(int)
```

```python
data.head()
```

```python
## Dropping The Null values
data.dropna(inplace = True)
```

```python
data.reset_index(inplace=True)
```

```python
data = data.drop(122).reset_index(drop=True)
```

```python
data.iloc[[122]]
```

```python
data.columns
```

```python
### Fix blank space in column names
data.columns = data.columns.str.strip()
```

```python
data.info()
```

```python
## Changing the required column data types

data[['day','month','year','Temperature','RH','Ws']] = data[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
objects = [feature for feature in data.columns if data[feature].dtypes=='O']
for i in objects:
    if i!='Classes':
        data[i] = data[i].astype(float)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.head()
```

```python
data.to_csv('Cleaned_data.csv',index=False)
```

```python
### Monthly Fire Analysis

plt.title("Fire Analysis of Sidi-Bel Abbes Region")
data['Classes'] = np.where(data['Classes'].str.contains('not fire'),'not fire','fire')
dtemp = data.loc[data['Region']==1]
sns.countplot(x='month',hue='Classes',data=dtemp,color='blue')
plt.xlabel("Month")
plt.ylabel('Count of Fires')
plt.figure(figsize=(30,25))
```

```python
plt.title("Fire Analysis of Bejaia Region")
dtemp = data.loc[data['Region']==0]
sns.countplot(x='month',hue='Classes',data=dtemp,color='blue')
plt.xlabel("Month")
plt.ylabel('Count of Fires')
plt.figure(figsize=(30,25))
```

August had the most number of forest fires for both the regions. Most of the fires happened in June, July and August and less happened in September.
